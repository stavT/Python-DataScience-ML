```python
### Pt. 1 Preparing the DATA ###

import pandas as pd
import numpy as np 
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('DATA/fake_reg.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values

y=df['price'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# help(MinMaxScaler)
```

```python
#only need to scale the features
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.max()
```

```python
X_train.min()
```

```python
X_train
```

```python
# X_train and X_test have been scaled to be from 0 to 1 
```

```python
### Pt. 2 Creating and Training the Model ###
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# model = Sequential([Dense(4,activation='relu'),Dense(2,activation='relu'),Dense(1)])
```

```python
#above and below are the same thing just done in different ways, methodology below allows for easier modifications

model = Sequential()

model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=250)
```

```python
model.history.history
# loss = model.history.history['loss']
```

```python
loss = model.history.history['loss']
sns.lineplot(x=range(len(loss)),y=loss)
plt.title("Training Loss per Epoch");
```

```python
### Pt. 3 Model Evaluation ###
```

```python
# This represents the loss that I decided the model to calculate which is mean squared error
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
test_predictions
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df.columns = ['Test True Y','Model Predictions']
```

```python
pred_df
```

```python
sns.scatterplot(x='Test True Y',y='Model Predictions', data=pred_df)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
# this means the model predictions error is about 4 dollars off from the true price point
```
